```python
import pandas as pd
import json

# Load base_data.csv
base_df = pd.read_csv('base_data.csv')

# Load csv_dictionary.json
with open('csv_dictionary.json', 'r') as f:
    csv_dict = json.load(f)

# Prepare a list to collect new rows
rows = []

# For each park in the dictionary
for park_code, records in csv_dict.items():
    for record in records:
        row = base_df[base_df['parkCode'].str.lower() == park_code.lower()].copy()
        if not row.empty:
            row = row.iloc[0].to_dict()
            row['year'] = record['Year']
            row['annual visits'] = record['RecreationVisitors']
            row['total visits'] = record.get('TotalRecreationVisitors', None)
            rows.append(row)

# Create the new DataFrame
combined_df = pd.DataFrame(rows)
combined_df.head()
combined_df.to_csv('final.csv', index=False)
```
